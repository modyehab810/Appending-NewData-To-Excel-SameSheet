### Create My Own DataFrame

```python
new_emplotees = {
    "Department": ["Developer", "Account Management", "Anlysis", "Developer", "Design", "Sales"], 
    "Status": ["Full Time", "Contracr", "Contract", "Full Time", "Contract", "Full Time"],
    "Hire_Date": ["16/11/2023", "10/10/2023", "09/10/2022", "29/18/2023", "21/05/2022", "11/04/2023"],
    "Years":[7, 4, 11, 10, 2, 4],
    "Benefits":["R", "DM", "M", np.nan, "R", "M"],
    "Compensation":[100000, 98000, 78950, 74510, 45000, 68000],
    "Job_Rating":[5, 4, 3, 5, 5, 3],
    "New_Comp.":[105000, 98500, 80500, 79000, 50000, 78000],
}
```

```python
new_df = pd.DataFrame(new_emplotees)
new_df
```

### Function That Append New DataFrame

```python
def append_to_excel_sheet(file_name, sheet, df):
    """
        Function That Append New DataFrame To Excel Sheet 
        • Take 3 Args (file_name, sheet, df)
        ► file_path: The Path Of The Excel File You Want To Append To it.
        ► sheet_name: The Name Of The Sheet To Append.
        ► df: The DataFrame You Want To Append.
        
    """
    seek = pd.read_excel(file_name, sheet_name = sheet).shape[0] + 1
    
    with pd.ExcelWriter(file_name, mode="a", if_sheet_exists="overlay") as writer:
        new_df.to_excel(writer, sheet_name = sheet, index=False, header=False, startrow=seek)    
    
```

```python
append_to_excel_sheet("employees_data.xlsx", "Sheet1", new_df)
```
